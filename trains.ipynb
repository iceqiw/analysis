```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


txt_data=''
with open('search.txt', 'r') as f:
    txt_data='[%s]' % ','.join(f.readlines())
    

original_data=pd.read_json(txt_data)
```

```python
original_data.head(10)
```

```python
data_group_date=original_data.groupby(['date'])
```

```python
train_sum=data_group_date.sum()

train_sum.head(10)
```

```python

train_sum['hard_sleeper'].plot(kind='bar')
plt.ylabel('hard_sleeper')
plt.show()
train_sum['hard_seat'].plot(kind='bar')
plt.ylabel('hard_seat')
plt.show()
train_sum['second_seat'].plot(kind='bar')
plt.ylabel('second_seat')
plt.show()
train_sum['soft_sleeper'].plot(kind='bar')
plt.ylabel('soft_sleeper')
plt.show()
train_sum.plot(use_index='date')
plt.show()
```
